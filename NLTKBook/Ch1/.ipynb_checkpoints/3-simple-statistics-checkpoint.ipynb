<h1>3 Computing with Language: Simple Statistics</h1>

```python
import nltk
from nltk.book import *
```

<h1>3.1 Frequency Distributions</h1>

<h3>Find the Frequency Distribution of Moby Dick</h3>

```python
# Find the frequency distribution of Moby Dick
fdist1 = nltk.FreqDist(text1)
print(fdist1, '\n\n')

# Find the fifty most frequent words
print(fdist1.most_common(50), '\n\n')

# Find how many occurances of whale
print(fdist1["whale"], '\n\n')
```

<h3>Cumulative Frequency Plot for the 50 Most Frequently Occuring Words in Moby Dick</h3>

```python
fdist1.plot(50, cumulative=True)
```

<h3>Hapaxes - Words that Only Occur Once</h3>

```python
hapaxes1 = fdist1.hapaxes()
print(hapaxes1)
```

<h1>3.2 Fine-Grained Selection of Words</h1>

<h3>Find All Words in the Book Longer than 15 Characters.</h3>

```python
V = sorted(set(text1))
long_words = [w for w in V if len(w) > 15]
print(long_words)
```

<h3>Find Words in the Book Longer Than 7 Characters that Occur > 7 Times</h3>

```python
frequent_long_words = [w for w in V if len(w) > 7 and fdist1[w] > 7]
print(frequent_long_words)
```

<h1>Collocations and Bigrams</h1>

<p>A collocation is a sequence of words that occur together unusually often. A characteristic of collocations is that they are resistant to substitution with words that have similar senses; for example, maroon wine sounds definitely odd.</p>

<h3>Extract Bigrams</h3>

```python
# Make a string
my_string = "this is a string that will be broken into bigrams"

# Break up the string into tokens
my_tokens = nltk.word_tokenize(my_string)

# Find all bigrams from the tokens
my_bigrams = list(nltk.bigrams(my_tokens))
print(my_bigrams)
```

<h3>Collocations</h3>
<p>Bigrams that occur more often than we would expect based on the frequency of the individual words. </p>

<p>e.g. Commonly occuring pairs of rare words.</p>

<h1>3.4 Counting Other Things</h1>

<h3>Distribution of Word Lengths</h3>

```python
# List of integer word lengths for all words in Moby Dick
t1_word_lengths = [len(w) for w in text1]

# Frequency Distribution of word lengths
fdist = FreqDist(t1_word_lengths)
fdist
```

<h3>Word Length Frequency Comparisons</h3>

```python
# Most common word lengths in order
print(fdist.most_common(), '\n\n')

# Most common word length
mcl = fdist.max()
print(mcl, '\n\n')

# Number of occurances of most common word length
mcl_count = fdist[mcl]
print(mcl_count, '\n\n')

# Relative frequency of the most common word length
mcl_freq = fdist.freq(mcl)
print(mcl_freq)
```
